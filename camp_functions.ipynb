```python
import numpy as np
import matplotlib.pyplot as plt
```

### Line plot

```python
# Create arrays (= lists which support math operations) of coordinates
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([3, 5, 9, -2, 1, 6])

# Plot
plt.plot(x, y)

# Add labels and titles
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Line plot")
```

### Scatter plot

```python
# Create arrays (= lists which support math operations) of coordinates
x = np.array([-2, 1.5, 2.4, 3.2])
y = np.array([0, -2, 9.0, 3.1])

# Plot
plt.scatter(x, y)

# Add labels and titles
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter plot")
```

### Plotting functions

```python
def line(x, slope, intercept):
    return intercept + x * slope

def parabola(x):
    return x ** 2
```

```python
# Generate 1000 points spaced uniformly between -5 and 5
x = np.linspace(-5, 5, 1000)

plt.plot(x, line(x, 3.0, 1.0), label = "Line")
plt.plot(x, parabola(x), label = "Parabola")
plt.legend()
```
