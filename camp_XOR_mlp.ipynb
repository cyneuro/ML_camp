## Logic Gates

XOR (exclusive OR) is a type of ***logic gate***.
<br>
Logic gates take two inputs (or more) and produce a boolean output, **true** ($1$) or **false** ($0$).
<br>

Let's assume any logic gates we use take inputs $A$ and $B$ ($0$ or $1$) and produce an output $Y$ ($0$ or $1$).

#### **AND**

The **AND** gate, as its name suggests, only returns $1$ when *both* inputs are $1$. If either or both of the inputs, $A$ and $B$, are $0$, then the gate outputs $0$.

#### **NAND and NOR**

**NAND** and **NOR** are short for ***NOT AND*** and ***NOT OR***, i.e. they are the *inverses* of AND and OR.
<br>
NAND — Anything with a $0$ as either input outputs $1$. If $A$ and $B$ are both $1$, $A$ NAND $B = 0$.
<br>
NOR — Anything with a $1$ as either input outputs $0$. If $A$ and $B$ are both $0$, $A$ NOR $B = 1$.

## Classifying logic gates

Take a look at the plots of XOR, AND, and OR below.

If you can recall, in the Logistic Regression tutorial we used one perceptron to separate data into two classes with a single line.

Is it possible to separate the data of each of these logic gates with one line?
<br>
If not, how many lines do you need?

```python
#@title
import matplotlib.pyplot as plt

# XOR
XOR_X = [[0, 0], [0, 1], [1, 0], [1, 1]]
XOR_y = [0, 1, 1, 0]

XOR_classes = {0 : [], 1 : []}
for x_i, y_i in zip(XOR_X, XOR_y):
  XOR_classes[y_i].append(x_i)

# AND
AND_X = [[0, 0], [0, 1], [1, 0], [1, 1]]
AND_y = [0, 0, 0, 1]

AND_classes = {0 : [], 1 : []}
for x_i, y_i in zip(AND_X, AND_y):
  AND_classes[y_i].append(x_i)

# OR
OR_X = [[0, 0], [0, 1], [1, 0], [1, 1]]
OR_y = [0, 1, 1, 1]

OR_classes = {0 : [], 1 : []}
for x_i, y_i in zip(OR_X, OR_y):
  OR_classes[y_i].append(x_i)

# Plotting

fig, axs = plt.subplots(1, 3)

# XOR
axs[0].set_title('XOR')
axs[0].set_xlabel('x1')
axs[0].set_ylabel('x2')
axs[0].scatter(*np.array(XOR_classes[0]).T, s=100, color='tab:blue', label='Class 0')
axs[0].scatter(*np.array(XOR_classes[1]).T, s=100, color='tab:red', label='Class 1')

# AND
axs[1].set_title('AND')
axs[1].set_xlabel('x1')
axs[1].set_ylabel('x2')
axs[1].scatter(*np.array(AND_classes[0]).T, s=100, color='tab:blue')
axs[1].scatter(*np.array(AND_classes[1]).T, s=100, color='tab:red')

# OR
axs[2].set_title('OR')
axs[2].set_xlabel('x1')
axs[2].set_ylabel('x2')
axs[2].scatter(*np.array(OR_classes[0]).T, s=100, color='tab:blue')
axs[2].scatter(*np.array(OR_classes[1]).T, s=100, color='tab:red')

for ax in axs:
  ax.grid()
  ax.set_aspect('equal')
  ax.set_xlim([-0.5, 1.5])
  ax.set_ylim([-0.5, 1.5])

plt.subplots_adjust(wspace = 0.75)
fig.legend(loc='upper center')

plt.show()
```

Let's try to use a single perceptron to classify each of these and see what happens.

```python
#@title **Challenge: What went wrong with the XOR plot and how can we fix it?**
from sklearn.linear_model import Perceptron
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

#------------------------------------------------#

# Plotting function

def plot_perceptron(model, classes, ax, title):
  # Plot dataset
  ax.scatter(*np.asarray(classes[0]).T, color='tab:blue', s=100)
  ax.scatter(*np.asarray(classes[1]).T, color='tab:red', s=100)

  # Plot decision boundary
  # Get θ-values from the model
  theta_0 = model.intercept_[0]
  theta_1, theta_2 = model.coef_[0]

  x_in = np.linspace(-0.5, 1.5, 50)
  y_pred = (-theta_1 * x_in - theta_0) / theta_2

  ax.plot(x_in, y_pred, color='tab:green')

  ax.set_title(title)
  ax.set_xlim([-0.5, 1.5])
  ax.set_ylim([-0.5, 1.5])
  ax.set_xlabel('x1')
  ax.set_ylabel('x2')
  ax.grid()
  ax.set_aspect('equal')

#------------------------------------------------#

# Data

# XOR
XOR_X = np.array(
        [[0, 0],
         [0, 1],
         [1, 0],
         [1, 1]])

XOR_y = np.array([0, 1, 1, 0])

XOR_classes = {0 : [], 1 : []}
for x_i, y_i in zip(XOR_X, XOR_y):
  XOR_classes[y_i].append(x_i)

# AND
AND_X = np.array(
        [[0, 0],
         [0, 1],
         [1, 0],
         [1, 1]])

AND_y = np.array([0, 0, 0, 1])

AND_classes = {0 : [], 1 : []}
for x_i, y_i in zip(AND_X, AND_y):
  AND_classes[y_i].append(x_i)

# OR
OR_X = np.array(
       [[0, 0],
        [0, 1],
        [1, 0],
        [1, 1]])

OR_y = np.array([0, 1, 1, 1])

OR_classes = {0 : [], 1 : []}
for x_i, y_i in zip(OR_X, OR_y):
  OR_classes[y_i].append(x_i)

#------------------------------------------------#

# Create a Perceptron and fit it to each dataset

model = Perceptron()

# Set up plot
fig, axs = plt.subplots(1, 3)

# Fit the perceptron to the XOR dataset and plot
model.fit(XOR_X, XOR_y)
plot_perceptron(model, XOR_classes, axs[0], 'XOR')

# Fit the perceptron to the AND dataset and plot
model.fit(AND_X, AND_y)
plot_perceptron(model, AND_classes, axs[1], 'AND')

# Fit the perceptron to the OR dataset and plot
model.fit(OR_X, OR_y)
plot_perceptron(model, OR_classes, axs[2], 'OR')

#------------------------------------------------#

# Adjust plot legend

plt.subplots_adjust(wspace = 0.75)

# Create a legend for the whole figure
legend_elements = [Line2D([0], [0], marker='o', color='w', markerfacecolor='tab:blue', markersize=10, label='Class 0'),
                   Line2D([0], [0], marker='o', color='w', markerfacecolor='tab:red', markersize=10, label='Class 1'),
                   Line2D([0], [0], color='tab:green', lw=2, label='Decision Boundary')]
fig.legend(handles=legend_elements, loc='upper center')

plt.show()
```

### Introducing Multilayered Perceptrons

Unlike AND and OR, XOR *cannot* be classified using one line.
<br>
No matter where you place the line, at least one point will be incorrectly classified.
<br></br>
However, what if we were able to draw *two* lines?
<br></br>
That's where the **Multilayered Perceptron (MLP)** comes in.
<br>
MLPs are exactly what they sound like: **layers of perceptrons that take inputs from the previous layer and output to the next layer, passing through each layer until a single output is found in the final layer.**
<br></br>
**Let's see what that looks like visually:**

A *single* perceptron, as shown in the AND and OR plots above, can only draw *one* line.

OnePerceptron.jpg
<br></br>
However, if we use the *combined* output of *two* perceptrons, we can classify the XOR problem with two lines.

In the diagram, for the sake of simplicity, each perceptron is condensed into a single circle ($h_1$, $h_2$) that takes in inputs ($x_0, x_1, x_2$) and computes an output ($\hat{y}$).

The weighted sum ($∑$) and sigmoid activation functions ($σ(z)$) for each perceptron won't be shown but still exist as part of the computation within the circle.

MLP.jpg
<br></br>
So, what exactly is happening in this MLP layout?

The model is divided into three layers: the ***input layer***, the ***hidden layer***, and the ***output layer***.

1. Both perceptrons, $h_1$ and $h_2$, receive data from the input layer ($x_0, x_1, x_2$). **Remember, $x_0$ is for *bias* and *always* equals $1$.**
2. $h_1$ and $h_2$ then compute a weighted sum ($z$) on the inputs with their respective weights ($\theta_0, \theta_1, \theta_2$), shown as colored arrows connecting perceptrons).
3. The perceptrons then *squash* their weighted sums with the sigmoid function ($\sigma(z)$), obtaining their outputs. We'll call them $\hat{y_1}$ and $\hat{y_2}$ for the outputs of $h_1$ and $h_2$, respectively.
4. The outputs of the hidden layer, $\hat{y_1}$ and $\hat{y_2}$, are then used as the *inputs* to the perceptron $o$ in the output layer.
5. The perceptron in the output layer, $o$, essentially acts as an AND gate to the outputs of the hidden layer, forming a *combined* boundary for classification based on the boundaries drawn by both $h_1$ and $h_2$.

In essence, perceptrons $h_1$ and $h_2$ both draw a decision boundary line like normal based on the data in the input layer.
<br>
The decision boundaries are then brought together in the output layer to form one unified solution.
<br></br>
The process of passing data into the input layer, through the hidden layers, and out of the output layer is known as **forward propagation** (the data propagates *forward* through the model).
<br></br>
**Let's try implementing this with sklearn!**

```python
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np

#----------------------------------------------#

# XOR Data

X = np.array(
        [[0, 0],
         [0, 1],
         [1, 0],
         [1, 1]])

y = np.array([0, 1, 1, 0])

#----------------------------------------------#

# Create MLP with 2 neurons in the hidden layer
model = MLPClassifier(hidden_layer_sizes=(2,), alpha=0.02, max_iter=8000, random_state=6)

# Fit the MLP to the data
model.fit(X, y)

#----------------------------------------------#

# Plotting

fig, ax = plt.subplots()

# Generate a grid over the input space to plot the color of the
# classification at that grid point
x_min, x_max = -0.5, 1.5
y_min, y_max = -0.5, 1.5

# Find model predictions over grid of entire plot
xx, yy = np.meshgrid(np.arange(x_min, x_max, .02),
                     np.arange(y_min, y_max, .02))
predictions = model.predict(np.c_[xx.ravel(), yy.ravel()])
predictions = predictions.reshape(xx.shape)

# Create a contour plot of the decision boundary
ax.contourf(xx, yy, predictions, cmap=plt.cm.Spectral_r, alpha=0.5)

# Plot the data points
ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral_r, edgecolors='k')

# Adjust the plot
legend_elements = [Line2D([0], [0], linestyle='', marker='o', color='w', markerfacecolor='tab:blue', markersize=10, label='Class 0'),
                   Line2D([0], [0], linestyle='', marker='o', color='w', markerfacecolor='tab:red', markersize=10, label='Class 1')]
ax.legend(handles=legend_elements, loc='upper left')
ax.set_title('XOR With MLP')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_aspect('equal')

# Display the plot
plt.show()

weights = model.coefs_
bias = model.intercepts_

print(weights)
print(bias)
```

Run the code below to plot XOR with noise. Noise may be added to classification problems to make the data more difficult to classify.

```python
#@title
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, FloatSlider

def plot_with_noise(noise):
  # Generate random XOR values
  A = np.random.choice([0, 1], size=100).astype(float)
  B = np.random.choice([0, 1], size=100).astype(float)
  Y = np.logical_xor(A, B)

  # Add noise to points
  A_noise = np.random.uniform(-noise, noise, 100)
  B_noise = np.random.uniform(-noise, noise, 100)
  A += A_noise
  B += B_noise

  # Create a plot
  fig, ax = plt.subplots()

  # Sort data by class
  classes = {False: [], True: []}
  data = zip((zip(A, B)), Y)
  for x_i, y_i in data:
    classes[y_i].append(x_i)

  # Plot data points
  ax.scatter(*np.array(classes[False]).T, color='tab:blue', label='Class 0 (false)')
  ax.scatter(*np.array(classes[True]).T, color='tab:red', label='Class 1 (true)')

  # Adjust plot
  ax.set_title('XOR')
  ax.set_xlabel('x1')
  ax.set_ylabel('x2')
  ax.set_xlim([-0.5, 1.5])
  ax.set_ylim([-0.5, 1.5])
  ax.set_aspect('equal')
  ax.legend(loc='upper right')
  ax.grid()

  # Display plot
  plt.show()

# Controls noise slider
_ = interact(plot_with_noise, noise=FloatSlider(value=0, min=0, max=0.5, step=0.01))
```

### Advanced - Updating Weights and Biases

Updating the parameters for each perceptron in an MLP is a more complicated process than for a single perceptron.  

However, it follows the same basic principle: **Calculate the gradients of loss with respect to each parameter, then update using the update rule**.

What complicates things, though, is the forward propagating nature of the model.

To compute gradients based on loss (only after an output is calculated), you have to trace your steps *backwards* through the model, calculating gradients in *reverse order*.
<br></br>
The process of calculating gradients *backwards* from the output layer to the first hidden layer is known as ***backpropagation***.
<br></br>
